2. Reduce Image Resolution:
    

*   NumPy reduces resolution by selecting every 3rd row and column
*   Print the new resolution of the image
*   Display the reduced-resolution image (By how much?)





```python
resized_photo = photo[::3,::3] #use another var and double colon for step size in both rows/columns
resized_height, resized_width = resized_photo.shape[:2]
print(f"Image Resolution Dimensions when NumPy selects every 3rd row and column: {resized_width} pixels x {resized_height} pixels")
print("")
width_reduction = ((width - resized_width)/width)*100
height_reduction = ((height - resized_height)/height)*100
print(f"Both Width reduced by {width_reduction:.2f}% and Height reduced by {height_reduction:.2f}% in pixel size")
print("")
plt.axis('off')
plt.imshow(resized_photo)
```

3. Flip the Image:

*   Horizontally: Flip the reduced-resolution image horizontally by reversing the column order
*   Vertically: Flip the reduced-resolution image vertically by reversing the order of rows.
*   Display both flipped images.




```python
print("This is the result of the image being flipped vertically:")
print("")
plt.axis('off')
plt.imshow(photo[::-1]) #vertically flipped
```

```python
print("This is the result of the image being flipped horizontally:")
print("")
plt.axis('off')
plt.imshow(photo[:,::-1]) #horizontally flipped
```

4. Convert the Image to Grayscale:

*   Convert the reduced-resolution image to grayscale using the formula:

Gray = 0.2989R + 0.5870G + 0.1140B
*   Display the grayscale image using plt.imshow() with the cmap = 'gray' option



```python
# altering by indices for R, G, B to be gray scale
Red = photo[:, :, 0]
Green = photo[:, :, 1]
Blue = photo[:, :, 2]

Gray = 0.2989 * Red + 0.5870 * Green + 0.1140 * Blue
print("This is how the image looks after applying Gray Scale function: ")
print("")

plt.imshow(Gray, cmap='gray')
plt.axis('off')
plt.show()
```

6. Brightness Adjustment:

*   Increase the brightness of the reduced-resolution image by adding 50 to the pixel values.
*   Use np.clip() to ensure pixel values remain in the range [0, 255].
*   Different brightness levels (e.g., -50 to darken, +100 to brighten)
*   Display the brightness-adjusted image.






```python
print("The image is brightened by adding 50 to each pixel:")
print("")
brightness_adjustment = 50 #this is how brightness is accounted for
bright_altphoto = np.clip(photo + brightness_adjustment, 0, 255) #now use np.clip to add restraints on pixel values
plt.axis('off')
plt.imshow(bright_altphoto)
```

```python
print("The image is darkened by subtracted 50 to each pixel:")
brightness_adjustment2 = -50
dark_altphoto = np.clip(photo + brightness_adjustment2, 0, 255)
plt.imshow(dark_altphoto)
plt.axis('off')
plt.show()
```
